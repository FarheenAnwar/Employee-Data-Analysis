## **EMPLOYMENT TENURE AND TURNOVER ANALYSIS**
Employment Tenure and Turnover Analysis involves examining the length of time employees stay with an organization and the rate at which employees leave the organization. This type of analysis helps in understanding employee retention, identifying trends in employee tenure, and calculating turnover rates.
* Distribution of Hire Dates
* Analysis of Employee Tenure
* Exit Date Analysis
* Employee Turnover Rate



### **Distribution of Hire dates**

```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, HTML

# Distribution of Hire Dates per Year
df['Hire Date'] = pd.to_datetime(df['Hire Date'])
hire_date_distribution = df['Hire Date'].dt.to_period('Y').value_counts().sort_index()

# Plotting the distribution of hire dates as a bar chart
plt.figure(figsize=(12, 6))
hire_date_distribution.plot(kind='bar', color='mediumseagreen', edgecolor='black')
plt.title('Distribution of Hire Dates')
plt.xlabel('Hire Date (Year)')
plt.ylabel('Number of Employees Hired')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

### **Analysis of Employee Tenure**

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
from IPython.display import display, HTML

# Calculate Employee Tenure
df['Hire Date'] = pd.to_datetime(df['Hire Date'])
df['Tenure'] = (datetime.now() - df['Hire Date']).dt.days / 365.25
tenure_distribution = df['Tenure']
print("\033[1;4mAnalysis of Employee Tenure:\033[0m")
print(tenure_distribution.describe())

# Plotting the tenure distribution as a histogram

plt.figure(figsize=(5,5))
tenure_distribution.plot(kind='hist', bins=20, color='cornflowerblue', edgecolor='black')
plt.title('Analysis of Employee Tenure')
plt.xlabel('Tenure (Years)')
plt.ylabel('Number of Employees')
plt.grid(True)
plt.show()

```

### **Exit Data Analysis**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Distribution of Exit Dates
df['Exit Date'] = pd.to_datetime(df['Exit Date'], errors='coerce')
exit_date_distribution = df['Exit Date'].dt.to_period('Y').value_counts().sort_index()
print("\033[1;4mExit Date Distribution:\033[0m")
print(exit_date_distribution)

# Plotting the distribution of exit dates as a line chart
display(HTML("<div style='text-align: center;'>"))  # Start centering
plt.figure(figsize=(12, 6))
exit_date_distribution.plot(kind='bar', color='darkorange',edgecolor='black')
plt.title('Exit Date Distribution')
plt.xlabel('Exit Date (Year)')
plt.ylabel('Number of Employees Exited')
plt.grid(True)
plt.show()

```

### **Employee Turnover Rate**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Calculate Turnover Rate
total_employees = df.shape[0]
exited_employees = df['Exit Date'].notna().sum()
turnover_rate = exited_employees / total_employees
print("\033[1;4mEmployee Turnover Rate:\033[0m")
print(f"Total Employees: {total_employees}")
print(f"Exited Employees: {exited_employees}")
print(f"Turnover Rate: {turnover_rate:.2%}")

# Data for Pie Chart
labels = ['Exited', 'Still Employed']
sizes = [exited_employees, total_employees - exited_employees]
colors = ['salmon', 'lightgreen']

# Plotting the turnover rate as a pie chart
plt.figure(figsize=(5,5))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Employee Turnover Rate')
plt.axis('equal')
plt.show()

```

## **PERFORMANCE AND REWARD ANALYSIS**
Performance and Reward Analysis involves examining the relationship between employee performance metrics and their rewards, such as salary, bonuses, and promotions. This analysis helps organizations understand if their reward systems are effectively incentivizing and recognizing employee performance.
* Correlation Between Annual Salary and Bonus Percentage
* Analysis of Bonuses by Job Title
* Analysis of Bonuses by Department



### **Correlation Between Annual Salary and Bonus Percentage**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Binning the Annual Salaries
salary_bins = [0, 50000, 100000, 150000, 200000, 250000, np.inf]
salary_labels = ['<50K', '50K-100K', '100K-150K', '150K-200K', '200K-250K', '>250K']
df['Salary Binned'] = pd.cut(df['Annual Salary'], bins=salary_bins, labels=salary_labels, right=False)

# Calculating average bonus percentage for each salary bin
salary_bonus_avg = df.groupby('Salary Binned')['Bonus %'].mean().reindex(salary_labels)

# Plotting the bar plot with enhancements
plt.figure(figsize=(7,7))
bars = plt.bar(salary_bonus_avg.index, salary_bonus_avg, color=plt.cm.viridis(np.linspace(0, 1, len(salary_bonus_avg))), edgecolor='black')

# Adding data labels
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.1f}%', ha='center', va='bottom', fontsize=10)

plt.title('Average Bonus Percentage by Salary Range')
plt.xlabel('Annual Salary Range')
plt.ylabel('Average Bonus Percentage')
plt.grid(True, axis='y')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

### **Analysis of Bonus by Job Title**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Grouped bar plot for analysis of bonuses by department
avg_bonus_by_dept = df.groupby('Department')['Bonus %'].mean().sort_values()

plt.figure(figsize=(5,5))
bars = plt.barh(avg_bonus_by_dept.index, avg_bonus_by_dept, color='skyblue', edgecolor='black')

plt.title('Average Bonus Percentage by Department')
plt.xlabel('Average Bonus Percentage')
plt.ylabel('Department')
plt.grid(True, axis='x')
plt.tight_layout()
plt.show()
```
